```python
# Generate a random regression problem
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html

import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
%matplotlib inline
X, y = make_regression(n_samples = 300, n_features=1, noise=5)
plt.scatter(X, y)
```

```python
# Generate a random (polynomial) regression problem
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html

import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
#%matplotlib inline
X, y = make_regression(n_samples = 300, n_features=1, noise=5)
y = y*y
plt.scatter(X, y)
```

```python
# Generate isotropic Gaussian blobs for clustering
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.figure(figsize=(4, 4))
X, y = make_blobs(n_samples=100, n_features=2, centers=3, cluster_std=1, random_state=3)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
```

```python
# Make a large circle containing a smaller circle in 2d.
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_circles.html

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
plt.figure(figsize=(8, 8))
X, y = make_circles(noise=0.2, factor=0.02, random_state=1)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
```

```python
# Make two interleaving half circles
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

plt.figure(figsize=(4, 4))
X, y = make_moons(n_samples=100, noise=0.1)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
```

```python
# Generate a random n-class classification problem
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# One informative feature, one cluster per class, two classes
X, Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1, n_classes=2)

#Two informative features, one cluster per class, two classes
#X, Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=2)

#Two informative features, two clusters per class, 2 classes
#X, Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=2, n_classes=2)

#Two informative features, one cluster per class, 3 classes
#X, Y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3)

plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, s=25, edgecolor='k')
```

```python
# Generate isotropic Gaussian and label samples by quantile
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_gaussian_quantiles.html

from sklearn.datasets import make_gaussian_quantiles
X1, Y1 = make_gaussian_quantiles(n_features=2, n_classes=3)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1, s=25, edgecolor='k')
```
