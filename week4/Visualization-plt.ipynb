```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# line plot
X = np.arange(10)
Y = 2*X*X*X + 3*X*X + 5*X 
plt.plot(X, Y, 'g--')# green dash line
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
# scatter plot
X = np.arange(10)
Y = np.random.randint(0, 100, len(X))
plt.scatter(X, Y, marker='x', c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.show()
```

```python
# pie plot
X = ('A', 'B', 'C', 'D', 'E')
Y = (10, 35, 20, 15, 20)
plt.pie(Y, labels=X)
#plt.title('pie plot')
plt.axis('equal')
#plt.show()
```

```python
# bar plot
X = ('A', 'B', 'C', 'D', 'E')
Y = (10, 35, 20, 15, 20)
plt.bar(x=X, height=Y, width=0.35)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('bar plot')
plt.show()
```

```python
# stacked bar plot
X = ('A', 'B', 'C', 'D', 'E')
Y1 = (10, 35, 20, 15, 20)
Y2 = (20, 10, 10, 30, 30)
plt.bar(x=X, height=Y1, bottom=None, width=0.35)
plt.bar(x=X, height=Y2, bottom=Y1,width=0.35)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('stacked bar plot')
plt.show()
```

```python
# subplots
plt.subplot(2, 2, 1)
X = ('A', 'B', 'C', 'D', 'E')
Y = (10, 35, 20, 15, 20)
plt.bar(x=X, height=Y, width=0.35)

plt.subplot(2, 2, 2)
X = ('F', 'G', 'H', 'I', 'J')
Y = (40, 35, 60, 15, 20)
plt.bar(x=X, height=Y, width=0.35, color='r')

plt.subplot(2, 2, 3)
X = ('K', 'L', 'M', 'N', 'O')
Y = (20, 35, 5, 35, 20)
plt.bar(x=X, height=Y, width=0.35, color='g')

plt.subplot(2, 2, 4)
X = ('P', 'Q', 'R', 'S', 'T')
Y = (5, 70, 50, 10, 20)
plt.bar(x=X, height=Y, width=0.35, color='y')

plt.show()
```
