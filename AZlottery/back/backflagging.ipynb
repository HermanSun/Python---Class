```python
import pandas as pd
import numpy as np
import math
#import statistics
from operator import itemgetter
```

```python
cost=pd.read_excel('TicketCost.xlsx', index_col=0)
```

```python
cost.head()
```

```python
cost["Ticket Cost"]=cost["Beginning Inventory Ticket Count"]/1000*cost["Scratcher Cost Per 1000"]
```

```python
cost.head()
```

```python
backlist=[688,692,818,901,959,1016,1018,1049,1072,1073,1167]
print(backlist)
```

```python
for i in cost.index:
    if i in backlist:
        cost.loc[i,"flag"]=1
    else:
        cost.loc[i,'flag']=0
```

```python
cost.head()
```

```python
cost['Ticket Revenue Margin']=cost["Scratcher Sales"]/cost["Ticket Cost"]-1
```

```python
cost.head()
```

```python
cost['Ticket Sale Rate']=cost['Scratcher Ticket Count']/cost['Beginning Inventory Ticket Count']
```

```python
cost['Total Cost']=cost['Ticket Cost']+cost['Validation Amount']
```

```python
cost['Net Revenue']=cost['Scratcher Sales']-cost['Total Cost']
```

```python
cost['Net Revenue Margin']=cost['Net Revenue']/cost['Total Cost']
```

```python
cost.to_excel("BackCost.xlsx")
```
